```python
import pandas
import serial
import datetime
```

```python
df = pandas.DataFrame(columns=('sl','time','light'))
arduino = serial.Serial('/dev/ttyACM1',9600) #Your port name
cnt=0
cout=0
while True:
    if arduino.inWaiting()>1:
        time = str(datetime.datetime.now())
        myData = arduino.readline()
        myData = float(myData)
        df.loc[cnt] = cnt,time,myData
        cnt = cnt+1
        print(cnt,'\n')
        if cnt==8000:
            df.to_csv('out'+str(cout)+'.csv',sep=',',encoding='utf-8') #Name of the CSV file
            cout = cout+1
            cnt = 0
        
```
